# Collections Merge 

In this notebook I will analyze two dataset collection obtained from the ACORDAR datasets.json files download. The two collection versions were downloaded one month after the other, so this demonstrate how the download process is not reproducible


```python
"""
@param path: path to the tree file of a collection
@return dictionary with key: dataset, value: list of files downloaded
"""
def buildDictionary(path: str) -> dict():
    list_file = open(path, "r")
    list_object = dict()

    list_file.readline()

    dataset = ""
    while True:
        line = list_file.readline()

        if not line:
            break

        if "dataset-" in line:
            dataset = line.split(" ")[1].strip("\n")
            list_object[dataset] = set()

        elif "dataset_metadata" not in line:
            split = line.split(" ")
            list_object[dataset].add(split[len(split) - 1].strip("\n").lower())

    return list_object
```

```python
riccardo_list_path = "/home/manuel/Tesi/ACORDAR/datasets_raw_riccardo.txt"
manuel_list_path = "/home/manuel/Tesi/ACORDAR/datasets_raw_manuel.txt"

manuel_list = buildDictionary(manuel_list_path)
riccardo_list = buildDictionary(riccardo_list_path)
```

```python
"""
@param riccardo_list content of riccardo collection
@param manuel_list content of manuel collection
@return dictionary with the files that are in manuel_collection and not in riccardo_collection
"""
def findDifferencesDatasets(riccardo_list:dict, manuel_list: dict) -> dict:
    differences = dict()
    for dataset, riccardo_files in riccardo_list.items():

        manuel_files = manuel_list[dataset]

        differences[dataset] = set()

        for file in manuel_files:
            if file not in riccardo_files:
                differences[dataset].add(file)
        
    return differences
```

```python
differences = findDifferencesDatasets(riccardo_list, manuel_list)
```

```python
for dataset, files_difference in differences.items():
    if len(files_difference) > 0:
        print("Dataset: "+dataset+" Diff: "+str(files_difference))
```

```python
# Dataset empty analysis

for dataset, riccardo_files in riccardo_list.items():
    manuel_files = manuel_list[dataset]

    if len(riccardo_files) == 0 and len(manuel_files) > 0:
        print("Empty Dataset: "+dataset+" Files: "+str(manuel_files))

```

```python
# insert the additional files in manuel collection to the riccardo collection 
# and track this files in a directory

import os
import shutil

path_collection_dif = "/media/manuel/Tesi/Collection_Dif"
path_manuel_collection = "/media/manuel/Tesi/Datasets_RAW_Manuel"
path_riccardo_collection = "/media/manuel/Tesi/Datasets"

for dataset, riccardo_files in riccardo_list.items():
    manuel_files = manuel_list[dataset]

    if len(riccardo_files) == 0 and len(manuel_files) > 0:
        print("Empty Dataset: "+dataset+" Files: "+str(manuel_files))

        diff_folder = path_collection_dif+"/"+dataset
        destination_folder = path_riccardo_collection+"/"+dataset
        source_folder = path_manuel_collection+"/"+dataset

        if not os.path.exists(diff_folder):
            os.mkdir(diff_folder)
        
        for file in manuel_files:

            file_name = file

            if not os.path.exists(source_folder+"/"+file_name):
                for file_in_folder in os.scandir(source_folder):
                    if file_in_folder.name.lower() == file:
                        file_name = file_in_folder.name

            shutil.copy(source_folder+"/"+file_name, destination_folder+"/"+file_name)
            shutil.copy(source_folder+"/"+file_name, diff_folder+"/"+file_name)

        
```
