# Downloader Log Analysis

In this notebook we are going to analyze the downloader errors reported in its log file.

## HTTP Error codes

We want to retrieve all the possible error codes that are returned by the downloader. 

```python
import os   
```

```python
#open the log file

dirname = os.path.dirname(__name__)
logs_path = os.path.join(dirname, '../download/logs/downloader_errors.log')

log_file = open(logs_path, "r", encoding="utf-8")

errors_codes = set()

while True:
    line1 = log_file.readline()

    if not line1:
        break

    line2 = log_file.readline()
    line3 = log_file.readline()
    line4 = log_file.readline()

    split = line3.split("Error: ")

    errors_codes.add(split[1].strip())
```

```python
for code in errors_codes:
    if len(code) < 20:
        print(code)
```

## HTTP Error Messages

We want to retrieve all the possible error messages that are returned by the downloader. 

```python
for code in errors_codes:
    if len(code) >= 20:
        print(code)
```
